```python
Contaminantes=['DO(mgO2/L)', 'CBODf (mgO2/L)', 'No(ugN/L)','Po (ugP/L)']
def area_influencia(dataframe,Contaminante):
    df = pd.pivot_table(dataframe,index='Escenario', columns='x(km)', values=Contaminante)
    test=df.T.sort_values('x(km)', ascending=False).reset_index()
    for i in np.arange(len(test.columns))[1::]:
        idx = test.loc[3::][test.columns[i]].sub(test.loc[0][i]).abs().idxmin()
        test.loc[idx-1:idx+1][test.columns[i]]
        if (test.loc[idx][test.columns[i]]>test.loc[0][i]) & (test.loc[idx+1][test.columns[i]]<test.loc[0][i]):
            df.loc[df.index==test.columns[i],'ar_infl']=8.5-(test.loc[idx]['x(km)']+test.loc[idx+1]['x(km)'])/2
        else:
            df.loc[df.index==test.columns[i],'ar_infl']=8.5-(test.loc[idx]['x(km)']+test.loc[idx-1]['x(km)'])/2
    fig=px.histogram(df[['ar_infl']].sort_values('ar_infl', ascending=False),x=df[['ar_infl']].index,y='ar_infl',marginal='box',title=Contaminante,).update_layout(
    yaxis_title="√Årea de influencia (Km)")
    return fig
    
```
