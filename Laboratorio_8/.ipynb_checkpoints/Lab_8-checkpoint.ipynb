Primero que todo, deben obtener un vector de caracter√≠sticas del atributo `history_text`, utilizando `Bag of Words`. En este atributo se presenta una breve descripci√≥n de la historia de cada uno de los personajes de ficci√≥n presentes en el dataset. 

Pero... antes de empezar, ¬øQue es `Bag of Words`?...

`Bag of Words` es un modelo de conteo utilizado en Procesamiento de Lenguaje Natural (NLP) que tiene como objetivo generar una representaci√≥n vectorial (vector de caracter√≠sticas en nuestro cas) para cada documento a trav√©s del conteo de las palabras que contienen. 

La siguiente figura muestra un ejemplo de `Bag of Words` en acci√≥n:

<p align="center">
  <img src="https://user.oc-static.com/upload/2020/10/23/16034397439042_surfin%20bird%20bow.png" width="500">
</p>

Como pueden ver, el modelo de `Bag of Words` no resulta tan complicado, ¬øpero c√≥mo lo aplicamos en python?. 

Como podr√°n darse cuenta del ejemplo anterior, para facilitar el conteo ser√° necesario transformar cada uno de los documentos en vectores, donde cada una de las posiciones posee un car√°cter. Este proceso es conocido como **tokenizaci√≥n** y lo podemos realizar de la siguiente forma:

````python
nltk.download('punkt')
````

````python
docs = ['The teacher rocks like a good rock & roll',
             'the rock is the best actor in the world']


docs_tokenizados = [word_tokenize(doc)  for doc in docs]
docs_tokenizados
````

Podemos mejorar un poco m√°s el proceso de tokenizaci√≥n agregando 

- Stemming:  Definimos Stemming como un algoritmo basado en reglas que transforma las palabras a una forma general. Un ejemplo de stemming, es el siguiente:
- Eliminaci√≥n de Stopwords: Eliminaci√≥n de palabras muy frecuentes que entorpecen la clasificaci√≥n (por ejemplo, el, la los, la, etc...)

<p align="center">
  <img src="https://devopedia.org/images/article/218/8583.1569386710.png" width="300">
</p>


````python
# Definimos algunas stopword que queremos que sean eliminadas
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords

stop_words = stopwords.words('spanish')

# Definimos un tokenizador con Stemming
class StemmerTokenizer:
    def __init__(self):
        self.ps = PorterStemmer()
    def __call__(self, doc):
        doc_tok = word_tokenize(doc)
        doc_tok = [t for t in doc_tok if t not in stop_words]
        return [self.ps.stem(t) for t in doc_tok]

# Inicializamos tokenizador
tokenizador = StemmerTokenizer()

# Creamos algunos documentos
docs = ['The teacher rocks like a good rock & roll',
        'the rock is the best actor in the world',
        'New York is a beautiful city']

# Obtenemos el token del primer documento
[tokenizador(doc) for doc in docs]
````

````python
# Comparaci√≥n con el caso anterior
docs_tokenizados = [word_tokenize(doc) for doc in docs]
docs_tokenizados
````

#### Al Estilo Scikit

Scikit implementa `bag of words` a trav√©s de la clase `CountVectorizer()` la cual contiene muchas opciones para mejorar la tokenizaci√≥n.

````python
bow = CountVectorizer(tokenizer= StemmerTokenizer())
df = bow.fit_transform(docs)

pd.DataFrame(df.toarray(), columns=bow.get_feature_names_out())
````

Una de las cosas m√°s interesantes que provee son el use de n-gramas, los cuales, en palabras simples, son conjuntos de n-palabras que se concatenan entre si y que se consideran como tokens separados. 

Pensemos en `Nueva York`. Cuando se tokeniza Nueva York, se generan dos tokens independientes que a simple vista no tienen relaci√≥n: `Nueva` `York`.
Al usar n-gramas (en un rango min=1,max=2) , generamos tanto `Nueva` y `York` como tambi√©n `Nueva York` como un token independiente.

````python
bow = CountVectorizer(tokenizer= StemmerTokenizer(), ngram_range=(1,2))
df = bow.fit_transform(docs)

pd.DataFrame(df.toarray(), columns=bow.get_feature_names_out())
````

De los resultados, podemos ver que generamos vectores de conteo para cada una de las palabras que conforman el corpus.  Un punto extra que se agrega en esta obtenci√≥n de frecuencias son los bigramas, que b√°sicamente son el conjunto de palabras de tama√±o de aparecen juntas en el texto.

**To-Do:**

**Respuesta:**

````python
#### C√≥digo aqu√≠ ####
column_text = 'history_text'
atributos_de_interes = ['intelligence_score', 'strength_score', 'speed_score', 'durability_score', 'power_score', 'combat_score']

preprocessing_transformer = ColumnTransformer(
    transformers=[
        ('bow', CountVectorizer(tokenizer=StemmerTokenizer(), ngram_range=(1, 2)), column_text),
        ('MinMaxScaler', MinMaxScaler(), atributos_de_interes)
        ])
````

Genere un Pipeline con las caracteristicas solicitadas en la secci√≥n 1.1, un selector de mejores features `SelectPercentile` con m√©trica `f_classif` y percentile=90 y un clasificador `MultinomialNB()` por defecto.

Luego, separe el conjunto de datos en un conjunto de entrenamiento y prueba, donde las etiquetas estar√° dado por el atributo `alignment`. 

Entrene el modelo y reporte el desempe√±o con un `classification_report`. ¬ø Nos recomendar√≠a predecir la alineaci√≥n de BatCow con este clasificador?.

Finalmente, compare el modelo entrenado con un modelo Dummy estratificado y responda: ¬øEl clasificador entrenado es mejor que el dummy que entrega respuestas al azar?

**To-do:**
- [ ] Realizar un pipeline con las caracteristicas solicitadas en 1.1, ejecutar holdout y aplicar un clasificador `MultinomialNB()`.
- [ ] Entrenar el pipeline, calcular el `classification_report` asociado y comentar los resultados.

**Respuesta:**

````python
#### C√≥digo aqu√≠ ####
multinomial_pipeline = Pipeline(
    steps=[
        ('Preprocessing', preprocessing_transformer),
        ('Selection', SelectPercentile(f_classif, percentile=90)),
        ('multinomialnb', MultinomialNB()),
    ]
)

y = df_comics.loc[:, 'alignment']
X = df_comics.loc[:, [column_text] + atributos_de_interes]
````

````python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42, shuffle=True, stratify=y)
````

````python
multinomial_pipeline.fit(X_train, y_train)
y_pred = multinomial_pipeline.predict(X_test)
print('Resultados Multinomial NB:')
print(classification_report(y_pred, y_test))
````

Los mejores resultados se observan en precision para la etiqueta Good, pero es evidente que existe un desbalance de clases y la performance macro del modelo es muy baja. 

El recall en las etiquetas que ten√≠an soportes para ser evaluadas fue similar, lo que quiere decir que el porcentaje de entidades presentes en las etiquetas reales que fue encontrado por este clasificador, tiene una proporci√≥n similar.

Al observar el resultado de F1, que es la combinaci√≥n entre precision y recall, se observa la misma tendencia que precision, debido al desbalance de clases.

````python
from sklearn.dummy import DummyClassifier
dummy_pipeline = Pipeline(
    steps=[
        ("preprocessing", preprocessing_transformer),
        ("dummy", DummyClassifier(strategy="stratified")),
    ]
)
````

````python
dummy_pipeline.fit(X_train, y_train)
y_pred = dummy_pipeline.predict(X_test)
print('Resultados DummyClassifier:')
print(classification_report(y_pred, y_test))
````

DummyClassifier ignora todas las features de entrada y genera salidas aleatorias como respuesta a las predicciones, lo que permite saber si el modelo que se est√° implementando es mejor que clasificar al azar. Como el MultinomialNB s√≠ toma en cuenta las features y por lo tanto adquiere el sesgo presente en el conjunto de entrenamiento, los resultados que tiene el modelo dummy en las clases minoritarias es mejor. El porqu√© de esto, es porque al usar MultinomialNB apenas se logran clasificar ejemplos en las clases minoritaias (dado el entrenamiento), pero como al modelo dummy no le importa el entrenamiento, clasificar alg√∫n ejemplo en la clase minoritaria es mero azar, pero como el desbalance en las clases es tal, ese azar es mejor que lo obteniedo por el modelo sesgado a la clase mayoritaria.

En cambio, si se observa la clase mayoritaria para la cual MultinomialNB s√≠ est√° preparado dado el entrenamiento, la performance del modelo aleatorio es peor.

A pesar de los resultados en las clases minoritarias, el modelo MultinomialNB sigue siendo mejor, por lo que se puede afirmar que el modelo s√≠ est√° aprendiendo.

````python
pipeline = Pipeline(
    steps=[
        ('Preprocessing', preprocessing_transformer),
        ('select', SelectPercentile(score_func=f_classif)),
        ('model', MultinomialNB()),
    ]
)
````

````python
pipeline.get_params()
````

````python
params = [
       # MultinomialNB + hiperpar√°metros
       {'Preprocessing__bow__ngram_range': [(1,1),(1,2),(1,3)],
        'select__percentile': [20, 40, 60, 80],
        'model': [MultinomialNB()],
        'model__alpha': [0.1, 0.5, 1]
       },
       # RandomForest + hiperpar√°metros    
       {'Preprocessing__bow__ngram_range': [(1,1),(1,2),(1,3)],
        'select__percentile': [20, 40, 60, 80],
        'model': [RandomForestClassifier(random_state=42)],
        'model__criterion': ['gini', 'entropy'],
       },
       # Regresi√≥n Log√≠stica + hiperpar√°metros
       {'Preprocessing__bow__ngram_range': [(1,1),(1,2),(1,3)],
        'select__percentile': [20, 40, 60, 80],
        'model': [LogisticRegression(random_state=42)],
        'model__C': [0.1, 0.5, 1],
       }
]
````
